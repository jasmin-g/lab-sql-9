Importing all necessary libraries & make connection to mysql

```python
import pandas as pd
import numpy as np
import pymysql 
from sqlalchemy import create_engine
from getpass import getpass 
```

```python
password = getpass()
```

```python
connection_string = 'mysql+pymysql://root:'+password+'@localhost/sakila'
engine = create_engine(connection_string)
```

## Q1 How many distinct (different) actors' last names are there?

```python
query = 'SELECT COUNT(DISTINCT last_name) AS number_of_unique_names FROM actor;'
data = pd.read_sql_query(query, engine)
data.head()
```

## Q2 Add an additional column day_type with values 'weekend' and 'workday' depending on the rental day of the week.

```python
query2 = 'SELECT *,DAYNAME(rental_date) AS day_of_the_week FROM rental'
data2 = pd.read_sql_query(query2, engine)
data2.head()
```

## Q3 Get all films with ARMAGEDDON in the title.

```python
query3 = 'SELECT * FROM film WHERE title LIKE "%%ARMAGEDDON%%"'
data3 = pd.read_sql_query(query3, engine)
data3.head()
```

## Q4 Get 10 the longest films

```python
query4 = 'SELECT * FROM FILM ORDER BY length DESC LIMIT 10'
data4 = pd.read_sql_query(query4, engine)
data4.head()
```

## Q5 How many films include Behind the Scenes content?

```python
query5 = 'SELECT COUNT(special_features) AS "Behind the Scenes" \
FROM film WHERE special_features LIKE "%%Behind the Scenes%%"'
data5 = pd.read_sql_query(query5, engine)
data5.head()
```

## Q6 Which kind of movies (rating) have a mean duration of more than two hours?

```python
query6 = 'SELECT rating, ROUND(AVG(length),2) as avg_length FROM film GROUP BY rating HAVING avg_length > 120'
data6 = pd.read_sql_query(query6, engine)
data6.head()
```

## Q7 Rank films by length (filter out the rows that have nulls or 0s in length column). In your output, only select the columns title, length, and the rank.

```python
query7 = 'SELECT title, length, DENSE_RANK() OVER (ORDER BY length DESC) AS my_rank \
FROM film WHERE (length IS NOT NULL) AND (length > 0)'
data7 = pd.read_sql_query(query7, engine)
data7.head()
```
